技术路线：Rank-BM25 + Jieba/HanLP + sklearn

```python
import jieba
from rank_bm25 import BM25Okapi
import numpy as np
import json
```

```python
NUM_CHAPTERS = 10
with open(f'first{NUM_CHAPTERS}_chunks.json', 'r', encoding='utf-8') as file:
    chunks = json.load(file)

# 此处可以添加专有名词
def chinese_tokenizer(text):
    tokens = jieba.lcut(text.replace('\n', ' '))
    # return list(tokens)
    # 过滤单字
    return [t for t in tokens if len(t.strip()) > 1]

corpus = [chunk['chunk_text'] for chunk in chunks]
tokenized_corpus = [chinese_tokenizer(doc) for doc in corpus]

bm25 = BM25Okapi(tokenized_corpus)
```

```python
def retrieve(query, top_k=5):
    tokenized_query = chinese_tokenizer(query)
    # 获取每个 chunk 的 BM25 分数
    scores = bm25.get_scores(tokenized_query)
    # 获取分数最高的索引
    top_n_indices = np.argsort(scores)[::-1][:top_k]
    
    results = []
    for idx in top_n_indices:
        results.append({
            "chpt_id": chunks[idx]["chpt_id"],
            "chunk_id": chunks[idx]["chunk_id"],
            "score": scores[idx],
            "text_snippet": chunks[idx]["chunk_text"]
        })
    return results

query = "杨家 敌对 家族 豪门"
# results = retrieve(query, top_k=5)
# for res in results:
#     print(f"Chapter ID: {res['chpt_id']}, Chunk ID: {res['chunk_id']}, Score: {res['score']}\n")
from bm25 import BM25Retriever
retriever = BM25Retriever(f"first{NUM_CHAPTERS}_chunks.json")
results = retriever.retrieve(query, top_k=5)
for res in results:
    print(f"Chapter ID: {res['chpt_id']}, Chunk ID: {res['chunk_id']}, Score: {res['score']}, Content: {res['text_snippet']}\n")
```
