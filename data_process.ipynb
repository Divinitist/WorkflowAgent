```python
file_path = '圣王.txt'
with open(file_path, 'r', encoding='utf-8') as f:
    file_str = f.read()
print(f'总字符数：{len(file_str)}（以UTF-8字符计）')
print(f'预览：{file_str[:500]}')
```

```python
file_lines = file_str.split('\n\n')
print(f'总自然段数（含标题行）：{len(file_lines)}')
print(f'自然段预览：{[line[:10] for line in file_lines[:5]]}')
chapters = []
chapter_cnt = 0
for line in file_lines:
    if line.startswith(f'第{chapter_cnt + 1}章'):
        chapter_cnt += 1
        chapters.append({
            'id': chapter_cnt,
            'title': line.lstrip(f'第{chapter_cnt}章').strip(),
            'lines': []
        })
    elif chapter_cnt > 0:
        chapters[chapter_cnt - 1]['lines'].append(line.strip())
print(f'总章节数：{len(chapters)}')
print(f'标题预览：{[chapter["title"] for chapter in chapters[:5]]}')
preview = "\n".join(chapters[-1]['lines'][:5])
print(f'内容预览：{preview}')
```

```python
NUM_CHAPTERS = 10
import json
file_path = f'first10.json'
with open(file_path, 'w', encoding='utf-8') as file:
    json.dump(chapters[:NUM_CHAPTERS], file, indent=4, ensure_ascii=False)
print(f'圣王前{NUM_CHAPTERS}章已写入\'{file_path}\'')
```

```python
chunks = []
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=700, chunk_overlap=150,
    separators=['\n\n', '\n', '。', '！', '？', ' '],
    keep_separator=True
)
for i in range(NUM_CHAPTERS):
    cur_chunks = text_splitter.split_text('\n'.join(chapters[i]['lines']))
    print(f'第{i + 1}章分割段数：{len(cur_chunks)}')
    for j, chunk in enumerate(cur_chunks):
        chunks.append({
            'chpt_id': i + 1,
            'title': chapters[i]['title'],
            'chunk_id': j,
            'chunk_text': chunk,
        })
with open(f'first{NUM_CHAPTERS}_chunks.json', 'w', encoding='utf-8') as file:
    json.dump(chunks, file, indent=4, ensure_ascii=False)
print(f'已写入 first{NUM_CHAPTERS}_chunks.json')
```
